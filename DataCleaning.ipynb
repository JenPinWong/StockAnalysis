```python
import pandas as pd
```

```python
database = pd.read_csv("Counter_Database.csv")
```

```python
df = pd.DataFrame({"Name":database["Name"], "Sector":database["Sector"], "Code":database["Code"], "Market":database["Market"],
                   "Mktcap(m)":database["Mktcap"], "Share_count(m)":database["Share_count"], "EPS":database["EPS"],
                   "PERatio":database["PERatio"], "ROE":database["ROE"]})
```

```python
def convert_to_million(col_name, df):
    for i in range(len(df[col_name])):
        if df[col_name][i][-1] == "b":
            print(df[col_name][i])
            df[col_name][i] = float(df[col_name][i][:-1]) * 1000

        elif df[col_name][i][-1] == "m":
            print(df[col_name][i])
            df[col_name][i] = float(df[col_name][i][:-1])

        
convert_to_million("Mktcap(m)", df)
convert_to_million("Share_count(m)", df)

```
